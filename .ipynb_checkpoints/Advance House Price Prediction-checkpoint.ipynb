# Feature Engineering 

### Replace missing value with a new label

```python
feature_categorical_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>1 and dataset[features].dtype == 'O']

def replace_nan_val(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna('Missing')
    return data


dataset=replace_nan_val(dataset,feature_categorical_nan)

## for Test data

dataset_test=replace_nan_val(dataset_test,feature_categorical_nan)
```

```python
dataset.head()
```

```python
dataset[feature_categorical_nan].isnull().sum()
```

### Handle Numerical Data missing value

```python
feature_numerical_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype != 'O']


for feature in feature_numerical_nan:
    print(feature,dataset[feature].isnull().sum())


## for Test Data

test_feature_numerical_nan=[feature for feature in dataset_test.columns if dataset_test[feature].isnull().sum()>1 and dataset_test[feature].dtype != 'O']
```

```python
for feature in feature_numerical_nan:
    median_nan=dataset[feature].median()
    dataset[feature+'_nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_nan,inplace=True)

print(dataset[feature_numerical_nan].isnull().sum())


## for Test Data

for feature in test_feature_numerical_nan:
    median_nan=dataset_test[feature].median()
    dataset_test[feature+'_nan']=np.where(dataset_test[feature].isnull(),1,0)
    dataset_test[feature].fillna(median_nan,inplace=True)
```

```python
dataset.head(20)
```

```python
## Temporal Variables (Date Time Variables)

for feature in temporal_feature: 
    if feature!='YrSold':
        dataset[feature]=dataset['YrSold']-dataset[feature]

dataset_test['SalePrice']=dataset['SalePrice']
dataset.head()
```

#### Skewed Numerical Data apply Log Transformation

```python
num_feature=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_feature:
    dataset[feature]=np.log(dataset[feature])


## for Test Data

for feature in num_feature:
    dataset_test[feature]=np.log(dataset_test[feature])
```

```python
dataset.head()
```

### Handling Rare Categorical Feature

```python
## Ignore entry which ocuure less than 1%

for feature in categorical_feature: 
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_val')

## for Test Data

for feature in categorical_feature: 
    temp=dataset_test.groupby(feature)['SalePrice'].count()/len(dataset_test)
    temp_df=temp[temp>0.01].index
    dataset_test[feature]=np.where(dataset_test[feature].isin(temp_df),dataset_test[feature],'Rare_val')
```

```python
dataset.head()
```

```python
for feature in categorical_feature: 
    lable_order=dataset.groupby(feature)['SalePrice'].mean().sort_values().index
    lable_order={k:i for i,k in enumerate(lable_order,0)}
    dataset[feature]=dataset[feature].map(lable_order)

## for Test Data
for feature in categorical_feature: 
    lable_order=dataset_test.groupby(feature)['SalePrice'].mean().sort_values().index
    lable_order={k:i for i,k in enumerate(lable_order,0)}
    dataset_test[feature]=dataset_test[feature].map(lable_order)
```

```python
dataset.head()
```
